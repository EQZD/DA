```python
df = df[df["Category"] != "1.9"]
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
df["reviews"] = df["reviews"].astype(int)
```

```python
df["installs"] = (
    df["installs"]
      .str.replace(",", "")
      .str.replace("+", "")
      .astype(int)
)
```

```python
df["price"] = (
    df["price"]
      .str.replace("$", "", regex=False)
      .astype(float)
)
```

```python
def size_to_mb(size):
    if size.endswith("M"):
        return float(size.replace("M", ""))
    elif size.endswith("k"):
        return float(size.replace("k", "")) / 1024
    else:
        return None

df["size_mb"] = df["size"].apply(
    lambda x: size_to_mb(x) if isinstance(x, str) else None
)
```

```python
df = df.drop_duplicates(subset="app")
```

```python
df = df.drop(columns=["size"])
```

```python
#part5 is_paid log_reviews 
- installs_per_rating – combined popularity and quality metric
```

```python
df["is_paid"] = df["price"] > 0
```

```python
import numpy as np

df["log_reviews"] = np.log1p(df["reviews"])
```

```python
df["installs_per_rating"] = df["installs"] / df["rating"]
```

```python
rating_anomalies = df[(df["rating"] < 0) | (df["rating"] > 5)]
rating_anomalies
```

```python
Q1 = df["reviews"].quantile(0.25)
Q3 = df["reviews"].quantile(0.75)
IQR = Q3 - Q1

reviews_anomalies = df[
    (df["reviews"] < Q1 - 1.5 * IQR) |
    (df["reviews"] > Q3 + 1.5 * IQR)
]

reviews_anomalies.head()
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

price_anomalies = df[
    (df["price"] < Q1 - 1.5 * IQR) |
    (df["price"] > Q3 + 1.5 * IQR)
]

price_anomalies.head()
```
